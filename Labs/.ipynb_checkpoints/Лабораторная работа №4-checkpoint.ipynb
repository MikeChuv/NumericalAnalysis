```python
import numpy as np
import matplotlib.pyplot as plt
import sys
from scipy.interpolate import interp1d
import pprint
```

```python
def sigma(phi, y, x):
    n = x.shape[0]
    s = sum( (phi(x) - y) ** 2 )
    return np.sqrt(s / n)
```

```python
def findPhi(dsx, dsy, m):
    # размер матрицы - m + 1
    A = np.zeros((m+1, m+1))
    b = np.zeros(m+1)
    for i in range(m+1):
        for j in range(m+1):
            A[i, j] = sum(dsx ** (i+j))
        b[i] = sum(dsy * dsx ** i)
    a = np.linalg.solve(A, b)
    #print(m, a)
    phi = lambda x: sum(ai * x ** i for i, ai in enumerate(a))
    return phi
```

```python
x = np.array([1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020], dtype = np.float64)
y = np.array([8.9, 9.6, 9.9, 10.3, 10.4, 10.3, 10.2, 10.7])
```

```python
for i in range(0, 7):
    phi = findPhi(x, y, i) # phi - функция
    s = sigma(phi, y, x)
    print(i,"\t", s)
```

```python
print(findPhi(x, y, 3)(2019))
```

```python
phi = findPhi(x, y, 3)
plt.plot(np.arange(1950, 2021), [phi(j) for j in range(1950, 2021)], label=f'$\Phi(x)$')
plt.legend()
plt.grid()
plt.plot(x, y, 'ro')
```

```python
def polyInter(x, y, t):
    m = x.shape[0]
#     A = np.zeros((m, m))
#     for i in range(m):
#         for j in range(m):
#             A[i, j] = x[i] ** j
#     a = np.linalg.solve(A, y)
#     l = lambda x: sum(ai * x ** i for i, ai in enumerate(a))
    e = np.eye(m)
    l = 0.0
    for i in range(m):
        p = 1
        xhelp = x[i] - x + e[i]
        for j in range(m):
            if i != j:
                p *= t - x[j]
        l += y[i] * p / np.prod(xhelp)
    return l
    
#l3 = polyInter(x, y)
print(polyInter(x, y, x))
```

```python
plt.plot(np.arange(1950, 2021), [polyInter(x, y, j) for j in range(1950, 2021)], label=f'$\Phi(x)$')
plt.legend()
plt.grid()
plt.plot(x, y, 'ro')
```

### Задача 4.2
Дана функция $y = f(x)$.Приблизить функцию  методом интерполяции, используя многочлен Лагранжа. Степень многочлена $N$ подобрать таким образом, чтобы максимальная величина погрешности на отрезке $\left[ a, b \right]$ не превышала заданной величины $\varepsilon$ Построить графики многочленов и графики погрешностей. Приблизить функцию методом интерполяции, указанным в индивидуальном варианте. Сравнить полученные результаты. 

$y = x sin(2-x)$ \
$a = 1$ \
$b = 4$ \
$\varepsilon = 0.001$

$P(x)$ - квадратичный сплайн с дополнительным условием: $y'(a) = P'(a)$

```python
def y(x):
    return x * np.sin(2 - x)
```

```python
def splineMatrix(x, y, n):
    der = np.sin(2-x[0])-x[0]*np.cos(2-x[0])
    A = np.zeros((3*n, 3*n))
    b = np.zeros(3*n)
    xhelp = np.zeros(3*n-3)
    derhelp = np.zeros(3*n-6)
    xsq = lambda t, k: np.roll(np.append(np.array([1, t, t**2]), xhelp), k)
    dery = lambda t, k: np.roll(np.append(np.array([0, 1, 2*t, 0, -1, -2*t]), derhelp),k)
    for i in range(n):
        A[2*i] = xsq(x[i], 3*i)
        b[2*i] = y[i]
    for i in range(n):
        A[2*i+1] = xsq(x[i+1], 3*i)
        b[2*i+1] = y[i+1]
    for i in range(0, n-1):
        A[i+2*n] = dery(x[i+1], 3*i)
    A[3*n-1] = np.append(np.array([0, 1, 2*x[0]]),xhelp)
    b[3*n-1] = der
    a = np.linalg.solve(A,b)
    a = a.reshape(n,3)
    return a
```

```python
def splineInterp(x_data, x, m):
    step = (x_data[-1] - x_data[0]) / (x_data.shape[0] - 1)
    pos = (x - x_data[0]) // step
    pos = np.int64(pos)
    #l = []
    if pos.size != 1:
        res = np.zeros(pos.shape[0])
        for idx, p in enumerate(pos): 
            #почему-то давал только последний m[p]
            #si = lambda t: sum(ai * t ** i for i, ai in enumerate(m[p])) 
            #l.append(si)
            # обработка правого края (иначе выход за пределы)
            p = (p - 1) if p == int((x_data[-1] - x_data[0]) // step) else p
            res[idx] = sum(ai * x[idx] ** i for i, ai in enumerate(m[p]))
    else:
        res = sum(ai * x ** i for i, ai in enumerate(m[pos]))
    return res
```

```python
a = 1.0
b = 4.0
n = 14
x_data = np.linspace(a, b, n+1)
y_data = y(x_data)
```

```python
m = splineMatrix(x_data, y_data, n)
xls = np.linspace(a, b, 100)
yInterp = splineInterp(x_data, xls, m)
yls = y(xls)
```

```python
rp = abs(yls - yInterp)
rl = abs(yls - [polyInter(xls, yls, j) for j in xls])
print(max(rp))
print(max(rl))
plt.plot(xls, rp)
plt.plot(xls, rl)
plt.legend(['spline', 'Lagrange'], loc='best')
```

```python
plt.figure(figsize=(12, 8), dpi=80)
plt.plot(xls, yInterp)
plt.plot(x_data, y_data, 'o')
plt.plot(xls, [polyInter(xls, yls, j) for j in xls])
plt.plot(xls, y(xls))
plt.legend(['spline','interp knots', 'Lagrange', 'f(x)'], loc='best')
```

### Задача 4.3
Задана функция $f(x)$, определенная на отрезке $\left[-1; 1 \right]$. Требуется разложить функцию в ряд Тейлора в окрестности нуля с точностью $\varepsilon$ и произвести экономизацию полученного степенного ряда.

$f(x) = e^{2x} - 1$

```python
def n_term_k(k):
    return 2 ** k / np.math.factorial(k)

def S(x, n):
    res = 0.0
    for i in range(1, n):
        res = res + (x ** i) * n_term_k(i)
    return res

def y(x):
    return np.exp(2*x) - 1

n = 1
x = np.linspace(-1, 1, 1000)
eps = 1e-4
while max(abs(S(x, n) - y(x))) >= eps:
    n += 1
print(n)
plt.plot(x, y(x))
plt.plot(x, S(x, n))
plt.grid()
plt.legend(["y(x)", "Taylor"], loc="best")
```

```python
def T(x, n):
    if n == 0:
        return 1
    if n == 1:
        return x
    return 2 * x * T(x, n-1) - T(x, n-2)
```

```python
def x11(x):
    return (1 / 1024) * (2816*x**9 - 2816*x**7 + 1232*x**5 - 220*x**3 + 11*x + T(x, 11))

def x10(x):
    return (1 / 512) * (1280*x**8 - 1120*x**6 + 400*x**4 - 50*x**2 + 1 + T(x, 10))
```

```python
delta_y = abs(y(x) - S(x, 10) - x11(x) * n_term_k(11))
delta_1 = max(delta_y)
print(delta_1)
plt.figure(figsize=(12, 8), dpi=80)
plt.grid()
plt.plot(x, delta_y)

delta_y = abs(y(x) - S(x, 9) - x10(x) * n_term_k(10))
delta_1 = max(delta_y)
print(delta_1)
plt.plot(x, delta_y)
#print(max(abs(y(x) - S(x, 10))))
plt.plot(x, abs(y(x) - S(x, 11)))
plt.plot(x, abs(y(x) - S(x, 10)))
plt.legend(['eco11', 'eco10','dS11', 'dS10'], loc='best')
```
