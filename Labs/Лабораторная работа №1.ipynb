## Поиск машинного нуля

```python
k = 10
e = 2 ** (-k)
while e != 0:
	k = k + 1
	e = 2 ** (-k)
k = k - 1
print(k)
print(2 ** (-k))
```

## Поиск машинной бесконечности

```python
k = 10.0
e = 2.0 ** k
try:
    while True:
        k = k + 1
        e = 2 ** k
    k = k - 1
except(OverflowError):
    print(k)
    print(2 ** (k - 1))
```

## Поиск машинного эпсилон

```python
def eps():
    k = 10
    e = 2 ** (-k)
    while 1 + e != 1:
        k = k + 1
        e = 2 ** (-k)
    k = k - 1
    return 2 ** (-k)

print(eps()) # различимое значение
```

# Задача 1.2 - вариант 33

```python
def F(x):
    return np.exp(1) ** x + np.cos(x)

def fact(k):
    if k == 0:
        return 1
    else:
        return k * fact(k - 1)

def n_term(x, k):
    return ((x ** k) / fact(k)) + ((-1) ** k) * (x ** (2 * k)) / fact(2 * k)

def S(x, N):
    res = 0.0
    for i in range (N):
        res = res + n_term(x, i)
    return res
```

```python
print(F(0.3))
print(S(0.3, 5))
```

```python
a = -1
b = 2
c = (a + b) / 2
x_data = np.linspace(a, b, 1000)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15,5))

axs[0].plot(x_data, F(x_data), label='$F$', color="black")

# стили линий
line_styles = ((0, (3, 1, 1, 1, 1, 1)), 'dotted', 'dashed', 'dashdot', (0, (1, 10)))

# функция и ряды
for i in range(1, 6):
    axs[0].plot(x_data, S(x_data, i), label=f'$S(x, {i})$', ls = line_styles[i - 1])

# разница
for i in range(1, 6):
    axs[1].plot(x_data, np.abs(S(x_data, i) - F(x_data)), label=f'$\Delta S(x, {i})$', ls=line_styles[i - 1])
    
axs[0].legend()
axs[1].legend()
```

```python
def abs_error(x, N):
    return np.abs(S(x, N) - F(x))
    
def rel_error(x, N):
    return abs_error(x, N) / np.abs(S(x, N))
```

```python
# разложение в окрестности нуля и 0 - центральная точка отрезка
p_sum = n_term(c, 0)
N_machine_error = 2
n_t = n_term(c, 1)
print(n_t / p_sum)
while np.abs(n_t / p_sum) > eps():
    p_sum += n_t
    N_machine_error += 1
    n_t = n_term(c, N_machine_error)
    print(n_t / p_sum)
print(N_machine_error)
```

```python
print(abs_error(c, 10))
print(rel_error(c, 10))
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15,10))

axs[0][0].plot(x_data, np.abs(F(x_data) - S(x_data, N_machine_error)),
               label=f'$\Delta S(x,{N_machine_error})$')
axs[0][1].plot(x_data, np.abs((F(x_data) - S(x_data, N_machine_error)) / S(x_data, N_machine_error)),
               label=f'$\delta S(x,{N_machine_error})$')
axs[1][0].plot(x_data, np.abs(F(x_data) - S(x_data, N_machine_error)), 
               label=f'$\Delta S(x,{N_machine_error})$')
axs[1][1].plot(x_data, np.abs((F(x_data) - S(x_data, N_machine_error)) / S(x_data, N_machine_error)),
               label=f'$\delta S(x,{N_machine_error})$')

axs[0][0].legend()
axs[0][1].legend()
axs[1][0].legend()
axs[1][1].legend()
axs[1][0].set(ylim=(0.0,10E-15))
axs[1][1].set(ylim=(0.0,10E-15))
```

```python
def ROUND(x):
    return float(np.format_float_scientific(x, precision=4))
```

```python
print(111111111, ROUND(111111111))
print(0.000000111111111, ROUND(0.000000111111111))
```

```python
def S_rounded_(x, N):
    res = 0
    for i in range(N):
        res += ROUND(n_term(x,i))
        res = ROUND(res)
    return res

def S_rounded(x, N):
    res = x.copy()
    for i in range(len(x)):
        res[i] = S_rounded_(x[i], N)
    return res
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

axs[0].plot(x_data, np.abs(F(x_data) - S_rounded(x_data, N_machine_error)), 
            label=f'$\Delta S(x,{N_machine_error})$')
axs[1].plot(x_data, np.abs((F(x_data) - S_rounded(x_data, N_machine_error)) / S_rounded(x_data, N_machine_error)),
            label=f'$\delta S(x,{N_machine_error})$')
axs[0].legend()
axs[1].legend()
```

**Вывод:** 
1. На первых двух графиках `Out[8]` видно, как с ростом количества слагаемых ряда тейлора уменьшается абсолютная погрешность.
2. На нижних графиках `Out[12]` можно увидеть колебания графика абсолютной и относительной погрешности в районе 0 - точки разложения, поскольку масштаб выбран на уровне машинного эпсилон, а также можно заметить, что при отдалении от центра отрезка в положительном направлении оси *x* растет абсолютная и отрицательная погрешность (отдаляемся от точки разложения и точки, в которой считали количество членов суммы, при которых относительная погрешность меньше машинного эпсилон)
3. Так как количество разрядов мантиссы при округлении равно 5, то погрешность имеет порядок $10^{-5}$, что можно видеть на графике `Out[16]`

## а) используя погрешности арифметических операций

$$f(a, b, c) = \frac{a + b}{b - c} $$
$$a = 52.31$$
$$b = 48.95$$
$$c = 47.81$$

$$\Delta (a) = \Delta (b) = \Delta (c) = 0.01 - \text{Оценка сверху}$$

```python
a_3 = 52.31
b_3 = 48.95
c_3 = 47.81

da3 = 0.01 / a_3
db3 = 0.01 / b_3
dc3 = 0.01 / b_3

print(da3)
print(db3)
print(dc3)
```

$$ \delta (a) = \frac{0.01}{52.31} = 0.00019 $$
$$ \delta (b) = \frac{0.01}{48.95} = 0.00020 $$
$$ \delta (c) = \frac{0.01}{47.81} = 0.00020 $$

$$ \delta(a + b) \leq \delta_{max} = \max{\delta(a), \delta(b)} = dab3$$
$$ \delta(b - c) \leq \delta_{max} * \left| \frac{b + c}{b - c} \right| = dbc3$$
$$ \delta(\frac{a + b}{b - c}) \leq \frac{dab3 + dbc3}{1 - dbc3} * dbc3 $$

```python
# погрешность суммы a + b
dab3_sum = max(da3, db3)
print("delta (a + b)")
print(dab3_sum, "\n")

# погрешность разности b - c
dbc3_sub = max(db3, dc3) * abs((b_3 + c_3)/(b_3 - c_3))
print("delta (b - c)")
print(dbc3_sub, "\n")

# погрешность частного
df1 = (dab3_sum + dbc3_sub) / (1 - dbc3_sub) * dbc3_sub
print("delta f")
print(df1, "\n")

# значение функции
fun1 = (a_3 + b_3) / (b_3 - c_3)
print("f")
print(fun1, "\n")

# абсолютная погрешность
Df1 = df1 * fun1
print("Delta f")
print(Df1, "\n")
```

$$ f(a, b, c) = (88.825 \pm 0.027)$$
$$ f(a, b, c) = 88.8246(1 \pm 0.0003)$$
$$ f(a, b, c) = 88.8$$

## б) используя общую формулу погрешности

$$f\left( a,b,c\right) =\frac{a+b}{b-c}$$

$$\frac{\partial f}{\partial a}=\dfrac{1}{b-c}$$
$$\dfrac{\partial f}{\partial b}=\dfrac{\left( b-c\right) -\left( a+b\right) }{\left( b-c\right) ^{2}}=\dfrac{-a-c}{\left( b-c\right) ^{2}}$$
$$\dfrac{\partial f}{\partial c}=\dfrac{a+b}{\left( b-c\right) ^{2}}$$
$$\Delta f\leq \dfrac{1}{b-c}\Delta \left( a\right) +\dfrac{-a-c}{\left( b-c\right) ^{2}}\Delta \left( b\right) +\dfrac{a+b}{\left( b-c\right) ^{2}}\Delta \left( c\right) $$

```python
Df2 = 0.01 * ( (1/(b_3 - c_3)) + ((-a_3-c_3)/(b_3 - c_3)**2) + ((a_3 + b_3)/(b_3 -c_3)**2) )
print(Df2)
```

```python
df2 = Df2 / fun1
print(df2)
```

$$ f(a, b, c) = (88.825 \pm 0.018)$$
$$ f(a, b, c) = 88.8246(1 \pm 0.0002)$$
$$ f(a, b, c) = 88.8$$

**Вывод:** С использованием общей формулы погрешность меньше, поскольку при вычислении по действиям, каждый этап вносит свою погрешность. Поэтому способ с использованием общей формулы дает более точный результат.
